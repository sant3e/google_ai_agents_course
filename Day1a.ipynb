## What is an AI Agent?

Think of an AI agent as a **smart assistant** that can:
- **Understand** your requests (like a chatbot)
- **Think** about how to respond (using an LLM)
- **Act** on your behalf (using tools like search, calculations, etc.)

### Simple Analogy

Imagine you ask a personal assistant: "What's the weather in London?"

1. **Understanding**: The assistant understands you want weather information
2. **Thinking**: They realize they need current weather data
3. **Acting**: They check a weather app or website
4. **Responding**: They tell you "It's 15Â°C and cloudy in London"

An AI agent does the same thing, but with code instead of a person!

## What is Google's ADK?

**ADK (Agent Development Kit)** is Google's framework for building AI agents. It provides:

- **Pre-built components** for common agent tasks
- **Integration** with Google's Gemini models
- **Tools** like Google Search, code execution, and more
- **Structure** for organizing complex agent behaviors

### Why use ADK instead of just calling an LLM directly?

| Direct LLM Call | ADK Agent |
|----------------|----------|
| Can only answer based on training data | Can search for current information |
| Limited to text responses | Can take actions (search, calculate, etc.) |
| No memory between calls | Can maintain conversation context |
| Manual prompt engineering | Structured agent architecture |

### Key Components Explained

1. **Agent**: The main component that contains the logic and instructions
2. **Model**: The LLM that powers the agent's thinking (e.g., Gemini)
3. **Instruction**: The prompt that tells the agent how to behave
4. **Tools**: Actions the agent can take (like searching the web)
5. **Runner**: The environment that executes the agent

## Creating Your First Agent

Let's create a simple agent that can answer questions using Google Search. We'll define its main properties:

````python
# Define our first agent with clear properties
root_agent = Agent(
    name="helpful_assistant",
    model="gemini-2.5-flash-lite",
    description="A simple agent that can answer general questions.",
    instruction="""You are a helpful assistant. 
    
    When answering questions:
    1. If the question is about current events or information after your training date, use Google Search
    2. If you're unsure about an answer, use Google Search to verify
    3. For general knowledge questions, you can answer directly
    4. Always provide clear, helpful responses
    
    Remember: You have access to Google Search for current information!""",
    tools=[google_search],
)

print("âœ… Root Agent defined successfully!")
print("ğŸ¤– Agent Properties:")
print(f"  â€¢ Name: {root_agent.name}")
print(f"  â€¢ Model: {root_agent.model}")
print(f"  â€¢ Tools: {[tool.__class__.__name__ for tool in root_agent.tools]}")
````

## Setting Up the Runner

The runner is what executes our agent. Think of it as the "engine" that powers our agent:

````python
# Create a runner for our agent
runner = InMemoryRunner(agent=root_agent)

print("âœ… Runner created successfully!")
print("ğŸƒ The runner is ready to execute our agent.")
print("\nğŸ’¡ What the runner does:")
print("  â€¢ Takes user input")
print("  â€¢ Passes it to the agent")
print("  â€¢ Executes the agent's logic")
print("  â€¢ Returns the agent's response")
````

## Testing Our Agent

Now let's test our agent with a question that requires current information:

````python
# Test the agent with a question about current weather
response = await runner.run_debug("What's the weather in London?")

print("ğŸ¤– Agent Response:")
print(response)
````
